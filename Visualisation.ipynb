```python
import pandas as pd
```

```python
data=pd.read_csv('advertising.csv')
```

```python
data.head(2)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

```python
data.dtypes
font={'size':28}
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(data.corr(),annot=True,linewidth=1,linecolor='w')
```

```python
plt.figure(figsize=(12,8))
plt.hist(data['Age'])
plt.hist(data['Age'][data['Clicked on Ad']==1])
plt.xlabel('Age',fontdict=font)
plt.ylabel('Clicked on Ad',fontdict=font)
```

```python
plt.figure(figsize=(12,8))
plt.hist(data['Male'])
plt.hist(data['Male'][data['Clicked on Ad']==1])
plt.xlabel('Male',fontdict=font)
plt.ylabel('Clicked on Ad',fontdict=font)
```

```python
plt.figure(figsize=(12,8))
plt.hist(data['Daily Internet Usage'])
plt.hist(data['Daily Internet Usage'][data['Clicked on Ad']==1])
plt.xlabel('Daily Internet Usage',fontdict=font)
plt.ylabel('Clicked on Ad',fontdict=font)
```
